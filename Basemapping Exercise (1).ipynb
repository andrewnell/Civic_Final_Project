```python
# Read in Building Footprints from NYC Open Data
footprints = pd.read_csv('rows.csv?accessType=DOWNLOAD')
#footprints = pd.read_json('https://data.cityofnewyork.us/resource/iiuk-ntxs.json')
footprints.head(2)
```

```python
# Clean data and drop irrelevant columns
footprints.drop(['bin','doitt_id',''], axis=1)
```

```python
plutoBK = pd.read_csv('BORO_zip_files_csv/BK.csv')
plutoBX = pd.read_csv('BORO_zip_files_csv/BX.csv')
plutoMN = pd.read_csv('BORO_zip_files_csv/MN.csv')
plutoQN = pd.read_csv('BORO_zip_files_csv/QN.csv')
plutoSI = pd.read_csv('BORO_zip_files_csv/SI.csv')
```

### Because PLUTO data comes separate for each borough, we're concatentating them here to construct a single PLUTO dataset for all of NYC

```python
# Concatenate PLUTO data from each of 5 borough into single dataset
pluto = pd.concat([plutoBK, plutoBX, plutoMN, plutoQN, plutoSI]).reset_index()
pluto.shape
```

```python
pluto.columns
```

```python
!wget https://data.cityofnewyork.us/api/views/ic3t-wcy2/rows.csv
```

```python
# Read in DOB Job Application Filings

DOBapps = pd.read_csv('rows.csv', dtype={'Block': object,'Lot': object})
DOBapps.shape

#, dtype={'block': str, 'lot': str}
```

```python
DOBapps.head(2)
```

```python
DOBapps.columns
```

```python
# Here we're mapping a digit to each borough name to replicate BBL format
DOBapps['num_boro'] = DOBapps['Borough'].map({'BRONX':'2','BROOKLYN':"3",'MANHATTAN':'1','STATEN ISLAND':'5','QUEENS':'4'})
DOBapps.tail(2)
```

```python
# Now we need to concatentae the borough, block, and lot numbes to create BBLs
DOBapps["BBL"] = DOBapps['num_boro']+DOBapps["Block"]+DOBapps["Lot"].str[1:]
DOBapps['BBL'].head(5)
```

### We have three datasets now (footprints, PLUTO, and permits); we want to merge them.

Building Footprints - footprints

PLUTO - pluto

DOB Construction Permits - DOBapps

```python
# Merging PLUTO data to Building Footprints
BF_merge = pd.merge(footprints, pluto, left_on='bbl', right_on='BBL')
```

```python
# BF_merge[["BBL","bbl"]].head(2)
```

```python
# Merging DOB permits to Building Footprints
three_way = pd.merge(BF_merge, DOBapps, on='BBL', how='outer')
```

```python
three_way.shape
```

```python
three_way.to_csv('Three_Way.csv')
```
